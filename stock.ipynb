```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import yfinance as yf

from datetime import date,datetime, time, timezone
```

```python
def get_stock_data(ticker,start,end):
  data = yf.download(tickers=[ticker], start = start, end = end)
  data.insert(0,"Ticker",ticker)
  return data
```

```python
ticker ='DIS'
start = datetime(2020,1,1)
end = datetime.today()
d = get_stock_data(ticker, start, end)
d.head()
```

```python
d = d.reset_index().pivot_table(
    index="Date",
    columns="Ticker",
    values="Close"
)

d.head()
```

```python
SPY = get_stock_data("SPY",start,end)
IYW = get_stock_data("IYW",start,end)
VT = get_stock_data("VT",start,end)
DBA = get_stock_data("DBA",start,end)
TLT = get_stock_data("TLT",start,end)
PDBC = get_stock_data("PDBC",start,end)
IAU = get_stock_data("IAU",start,end)
```

```python
SPY.info()
```

```python
SPY=SPY.reset_index().pivot_table(index="Date", columns="Ticker", values="Close")
IYW=IYW.reset_index().pivot_table(index="Date", columns="Ticker", values="Close")
VT=VT.reset_index().pivot_table(index="Date", columns="Ticker", values="Close")
DBA=DBA.reset_index().pivot_table(index="Date", columns="Ticker", values="Close")
TLT=TLT.reset_index().pivot_table(index="Date", columns="Ticker", values="Close")
PDBC=PDBC.reset_index().pivot_table(index="Date", columns="Ticker", values="Close")
IAU=IAU.reset_index().pivot_table(index="Date", columns="Ticker", values="Close")
```

```python
stock = pd.concat([SPY, IYW,VT,DBA,TLT,PDBC,IAU],axis=1,join='outer')
stock.head()
```

```python
plt.style.use('ggplot')
stock.plot(figsize = (20,10))
plt.show()
```

```python
covid = stock['2020-2-1':'2020-7-31']

import pandas as pd
import matplotlib.pyplot as plt

# Flatten MultiIndex columns
stock.columns = stock.columns.get_level_values(0)

# Ensure datetime index
stock.index = pd.to_datetime(stock.index)

# Verify columns
print(stock.columns)

# Slice COVID period
covid = stock.loc['2020-02-01':'2020-07-31']

# Check result
print(covid.shape)
print(covid.head())

# Plot
plt.style.use('ggplot')

covid.plot(figsize=(20,10))
plt.title('Asset Performance During COVID Period')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
